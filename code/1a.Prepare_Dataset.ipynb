```python
##Target Dataset 2017-07-27 thru 2020-2-1 
gw = pd.read_csv("./data/ManualDownload/Daily_GW.CSV",
                    parse_dates=['Datetime'],
                    index_col=['Datetime'])
stage= pd.read_csv("./data/ManualDownload/Stage_Daily.CSV",
                    parse_dates=['Datetime'],
                    index_col=['Datetime'])
spring=pd.read_csv("./data/ManualDownload/DailyWakulla.CSV",
                    parse_dates=['datetime'],
                    index_col=['datetime'])
```

```python
print(stage.shape)
stage=stage.resample('1D').mean()
print(stage.shape)
```

```python
gw_zoom=gw.loc['20170127':'20200201']
print(gw_zoom.isna().sum().sum())
gz=gw_zoom.interpolate()
print(gz.isna().sum().sum())
g=gz.apply(lambda x: x.fillna(x.mean()),axis=0)
print(g.isna().sum().sum())
g
```

```python
stage_zoom=stage.loc['20170127':'20200201']
print(stage_zoom.isna().sum().sum())
sz=stage_zoom.interpolate()
print(sz.isna().sum().sum())
s=sz.apply(lambda x: x.fillna(x.mean()),axis=0)
print(s.isna().sum().sum())
s
```

```python
wak=spring['Gage_Height']
wak
```

```python
gw_stage_inputs = pd.concat([g, s], axis=1)
gw_stage_inputs.isna().sum().sum()
#gw_stage_inputs
gs=gw_stage_inputs.interpolate()
gs.isna().sum().sum()
gs
```

```python
gs.to_csv("./data/processed/gw_stage.CSV")
wak.to_csv("./data/processed/spring.CSV")
```
